## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
```

# Dataset Upload

```
dataset = pd.read_excel('solar_ds.xlsx')
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
y
```

## Plotting graph


```
sb.relplot(x='AirTemperature',y='Sunshine',data=dataset)
```

```
sb.relplot(x='RelativeAirHumidity',y='Sunshine',data=dataset)
```

```
sb.relplot(x='WindSpeed',y='Sunshine',data=dataset)
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
print(y)
```

## Training the Multiple Linear Regression model on the Training set

```
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(X_train, y_train)
```

# KNN

```
from sklearn import neighbors
knn=neighbors.KNeighborsRegressor(metric='euclidean',n_neighbors=23)
knn.fit(X_train,y_train)
```

## Predicting the Test set results

```
y_pred = regr.predict(X_test)
y_pred
```

# KNN Prediction

```
y_pred1=knn.predict(X_test)
y_pred1
```

## Evaluating the Model Performance

```
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
